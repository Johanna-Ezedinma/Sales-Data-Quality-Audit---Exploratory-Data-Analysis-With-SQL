```python
%%sql
SELECT *
FROM sales_data
LIMIT 10;
```

**FIXING MISSING TERRITORIES**

```python
%%sql
UPDATE sales_data 
SET territory = 'AMER' 
WHERE country IN ('USA', 'Canada') 
AND (territory IS NULL OR territory = 'NA')
;
```

*confirmation*

```python
%%sql
-- CONFIRMATION

SELECT country, territory, COUNT(*) 
FROM sales_data 
WHERE territory IS NULL OR territory = 'NA'
GROUP BY country, territory;
```

**SAVING PROCESSED CSV VERSION**

The <<  saves the result into a variable called customers_result instead of just printing it.

index=False stops pandas from adding its own extra row-number column into the CSV, since table already has real ID columns.

```python
%%sql
customers_result << SELECT * FROM customers;
```

```python
# Convert to a pandas DataFrame, then save it
customers_df = customers_result.DataFrame()
customers_df.to_csv('../data/processed/customers.csv', index=False)
```
